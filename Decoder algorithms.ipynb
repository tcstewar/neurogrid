This notebook compares classic and classic2 decoder algorithms. The latter is much better.  It computes decoders by putting the noise along the diagonal of $\Gamma$ instead of in the $A$ matrix directly.

```python
import neurogrid as ng
import numpy as np

#Define the neuron tuning curve... ng or this will be the same overall effect (see later for ng neuron version)
def rectLin2(num, xSize):
    maxRange=[100,200]
    xRange=[-.95,.95] #Smaller than diameter (i.e 2*R)
    Radius = 1
    x = linspace(-Radius,Radius,xSize)
    
    xInts = (xRange[1]-xRange[0])*rng.rand(1,num)+xRange[0]  #Generate random x intercepts
    maxFRs = (maxRange[1]-maxRange[0])*rng.rand(1,num)+maxRange[0]  #Generate random max firing rates
    slopes = maxFRs/(Radius-xInts)  #Compute the slopes
    biases = -(slopes*xInts)
    
    slopes[0,0:floor(num/2.0)]=-slopes[0,0:floor(num/2.0)] #Flip half the neurons
    
    A = outer(slopes,x).T+biases
    return x, A.clip(min=0)
```

```python
rng = np.random.RandomState()
numNeurons=2**arange(0,10,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 25
numPowers = arange(0,6)

PavgMSE=[]
for P in numPowers:
    avgMSE=[]
    for N in numNeurons:
        MSE=[]
        for i in arange(numRuns):
            X, A = rectLin2(N,100) #The sample points can be ~100 for classic 2, but should be about 2*N or more for classic...
            #d = ng.decoders.classic(A.T, X**P, rng, minSV=0)
            d = ng.decoders.classic2(A.T, X**P, rng, minSV=0)  #This gives way better decoders with way fewer sample points
        
            #A = A + rng.randn(*A.shape)*(0.1*np.max(A))  #Either with or without noise, classic2 is better
            Xhat = np.dot(A, d)     #Compute estimate with decoders
            MSE.append(mean((X**P-Xhat)**2))  #Compute MSE
        
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt    
tmp=plt.plot(array(PavgMSE).T)
plt.legend(tmp,numPowers, loc=3)
plt.yscale('log')
plt.xticks(arange(0,11,1),array(numNeurons).astype('str'))
plt.plot(.1/numNeurons,'--')
plt.plot(.1/numNeurons**.5,'.')
plt.plot(.1/numNeurons**1.5,'*')
plt.plot(.1/numNeurons**2,'+')
disp('Accuracy for Nth Degree Polynomials with increasing neurons')
```

Same deal for neurogrid neurons.

```python
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,11,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 25
numPowers = arange(0,6)

PavgMSE=[]
for P in numPowers:
    avgMSE=[]
    for N in numNeurons:
        MSE=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=100) #Standard NEF decoders, sample count must be much higher for non-classic2 decoders
            d = ng.decoders.classic2(A, X**P, rng, minSV=0)
        
            Xhat = np.dot(A.T, d)     #Compute estimate with decoders
            MSE.append(mean((X**P-Xhat)**2))  #Compute MSE
        
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt    
tmp=plt.plot(array(PavgMSE).T)
plt.legend(tmp,numPowers, loc=3)
plt.yscale('log')
plt.xticks(arange(0,11,1),array(numNeurons).astype('str'))
plt.plot(.1/numNeurons,'--')
plt.plot(.1/numNeurons**.5,'.')
plt.plot(.1/numNeurons**1.5,'*')
plt.plot(.1/numNeurons**2,'+')
disp('Accuracy for Nth Degree Polynomials with increasing neurons')
```
