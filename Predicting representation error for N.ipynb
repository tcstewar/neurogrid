```python
import numpy as np
import matplotlib.pyplot as plt    
%pylab inline
```

```python
#Define rectified linear neuron rate function (scalars only)
def Gaussian(num, xSize, sigma, Radius):
    #xRange=[-.95,.95] #Smaller than diameter (i.e 2*R)
    x = linspace(-Radius ,Radius, xSize)
    means = linspace(-Radius, Radius, num)
    
    A = exp(-(x[:,None]-means[:])**2/(2*sigma**2))
    A = 1/(np.sqrt(2*np.pi*sigma**2)) * A
    return x, A
```

```python
def ComputeTk(k, sigma, delta, N, noise):
    sd = sigma/(np.sqrt(2)*delta)
    norm_factor = N/(4*delta*np.sqrt(2*np.pi)*sd)
    tk = norm_factor*np.exp(-k**2/(2*sd**2)) + N*noise**2
    return tk
```

```python
#The numerical routine as a function
def ComputeErrors(N):
    #N is Number of neurons
    sd = .05 #SD of gaussians
    Samples = 2000 #Number of samples of gaussian
    sigma = .1 #Noise term
    Radius = 1 #-1,1
    noise = True #Use noise or not
    
    X, A = Gaussian(N, Samples, sd, Radius)
    Aorig = np.array(A)
    
    if noise == True:
        sigma = .1*np.max(A) #Noise term scaled by curve height
    else:
        sigma = 0
    
    A = A + sigma**2*numpy.random.random(A.shape)
    
    #func = X**3 #Function to estimate
    func = X*(X-1)*(X+1)
    
    V,S,U = np.linalg.svd(A)
        
    #Compute decoders from S
    num_S = N
    Sinv = np.array(S)
    Sinv[num_S:]=0 
    Sinv[Sinv>0]=1.0/Sinv[Sinv>0]
    Uinv = np.dot(U.T,np.diag(Sinv))
    Vf = np.dot(V.T,func)
    decoders = np.dot(Uinv,Vf[0:N])
    
    decEst = np.dot((Aorig + sigma**2*numpy.random.random(A.shape)), decoders)
    E_empirical = np.sqrt(np.mean((func-decEst)**2)) #RMSE
    
    ##Theory
    C = np.fft.fft(func) #Project the function onto the fourier basis
    
    Tk = []
    for k in range(0, Samples): 
        Tk.append(ComputeTk(k-Samples/2.0, sd, 1.0/Samples,N, sigma))
    
    #Fourier transform of Tk estimate singular values
    Sa = np.real(np.fft.fft(np.fft.fftshift(Tk)))
    #Sa = np.repeat(Sa[0:Samples/2+1],2)[1:] #Duplicate eigs to match SV indexing
    #Sa = Sa + (N/2)*sigma**2  #Floor from variance of noise (not noise var)

    #plt.semilogy(Sa[0:50])
    #plt.semilogy((S**2)[0:50])
    
    #Compute the estimate with fourier basis
    E_total = np.real(np.fft.ifft((Samples*sigma**2/(Sa + Samples*sigma**2))*C))
            
    E_theory = np.sqrt(np.mean(E_total**2))  #RMSE
    
    return E_empirical, E_theory 
```

```python
#Generate data
Neurons = [2,4,8,16,32,64,128,256,512,1024]

emp = np.empty((len(Neurons),5))
theory = np.empty((len(Neurons),5))
for i in range(len(Neurons)):
    for j in range(5): #Run each 5 times, though this doesn't matter for the same curves all the time
        emp[i,j],theory[i,j]=ComputeErrors(Neurons[i])
```

```python
emp_mean = np.mean(emp,1)
theory_mean = np.mean(theory,1)
#Plots
#Plot actual and ideal singular values
fig, axes = subplots(1,2,figsize=(16,4))
axes[0].semilogy(emp_mean)
axes[0].semilogy(theory_mean)
axes[0].semilogy(1.0/np.array(Neurons),'r:')
axes[0].set_title('Empirical and theoretical error')
axes[0].legend(['Empirical', 'Theoretical', '1/N'])
```
