This notebook contains background material for predicting representation/computation error theoretically. It has 3 sections combining various kinds of projection, noise assumptions, etc.  A bunch of this background is not used in the final notebook, which is what you should really look at. It's called: Predicting representation error for N

```python
import numpy as np
import matplotlib.pyplot as plt    
%pylab inline
```

```python
#Define rectified linear neuron rate function (scalars only)
def Gaussian(num, xSize, sigma, Radius):
    #xRange=[-.95,.95] #Smaller than diameter (i.e 2*R)
    x = linspace(-Radius ,Radius, xSize)
    means = linspace(-Radius, Radius, num)
    
    A = exp(-(x[:,None]-means[:])**2/(2*sigma**2))
    A = 1/(np.sqrt(2*np.pi*sigma**2)) * A
    return x, A
```

## 'Pure' analytic version of error prediction

```python
#Use this to compare analytic and empirical error, assuming a fourier basis

sd = .05 #SD of gaussians
N = 200 #Number of neurons
Samples = 500 #Number of samples of gaussian
sigma = .1 #Noise term
Radius = 1 #-1,1

X, A = Gaussian(N, Samples, sd, Radius)

#func = X**3 #Function to estimate
func = X*(X-1)*(X+1)

V,S,U = np.linalg.svd(A)

C = np.fft.fft(func) #Project the function onto the A basis
spec = np.fft.fftshift(C)
spec = np.abs(spec)**2

fig, axes = subplots(1,3,figsize=(12,4))
axes[0].plot(spec)
axes[1].plot(np.fft.ifft(C))
axes[1].plot(func)
axes[2].plot(np.fft.fftshift(C))

num_C = N
C_static = np.array(C)
C_static[num_C/2:Samples-num_C/2]=0
C_static[0:num_C/2]=1
C_static[Samples-num_C/2:]=1
C_static = C_static*C  #Mask C
func_est = np.real(np.fft.ifft(C_static))

fig, axes = subplots(1,3,figsize = (12,4))
axes[0].plot(np.abs(np.fft.fftshift(C_static))**2)
axes[1].plot(func_est)
axes[1].plot(func)
axes[2].plot(np.fft.fftshift(C_static))

Sinv = np.array(S)
Sinv[N:]=0
Sinv[Sinv>0]=1.0/Sinv[Sinv>0]
Uinv = np.dot(U.T,np.diag(Sinv))
Vf = np.dot(V.T,func)
decoders = np.dot(Uinv,Vf[0:N])

dec_est = np.dot(A, decoders)

error = np.sqrt((func-dec_est)**2)
E_static = np.sqrt((func-func_est)**2)

fig, axes = plt.subplots(1,3,figsize=(12,4))
axes[0].plot(dec_est)
axes[0].plot(func)
axes[1].plot(error)
axes[2].plot(E_static)

print "decoder error", mean(error)
print "analytic error", mean(E_static)
```

## Numerical version of error prediction (no noise)

```python
#This uses the numerical bases to do the same kind of thing as the previous sectin
sd = .05 #SD of gaussians
N = 200 #Number of neurons
Samples = 1000 #Number of samples of gaussian
sigma = .1 #Noise term
Radius = 1 #-1,1

X, A = Gaussian(N, Samples, sd, Radius)

func = X**3 #Function to estimate

V,S,U = np.linalg.svd(A)

C = np.dot(V.T,func) #Project the function onto the A basis
fig, axes = subplots(1,2,figsize=(12,4))
axes[0].plot(C)
#axes[0].semilogy(S)
axes[1].plot(np.dot(V,C))
axes[1].plot(func)

C_zeroed = np.array(C)
C_zeroed[70:]=0
funcCest = np.dot(V,C_zeroed)

fig, axes = subplots(1,2,figsize = (12,4))
#plt.plot(V.T[110:120,:].T)
axes[0].plot(C_zeroed)
axes[1].plot(funcCest)
axes[1].plot(func)

E_static = np.dot(V[:,N+1:],C[N+1:,None])

errorC = np.sqrt((func-funcCest)**2)

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].plot(errorC[N+1:])
axes[1].plot(np.sqrt(E_static**2))

Gamma = np.dot(A.T,A)
Sinv = np.array(S**2)
Sinv[70:]=0
Sinv[Sinv>0]=1.0/Sinv[Sinv>0]
GammaInv = np.dot(np.dot(U.T,np.diag(Sinv)),U)
decoders = np.dot(np.dot(GammaInv,A.T),func)

funcEst = np.dot(A, decoders)
errorDec = np.sqrt((func-funcEst)**2)

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].plot(errorDec[N+1:])
axes[1].plot(funcEst)
axes[1].plot(func)
```

## Numeric prediction of noise error also

```python
#This part adds noise to the previous section
sd = .05 #SD of gaussians
N = 200 #Number of neurons
Samples = 1000 #Number of samples of gaussian
Radius = 1 #-1,1

X, A = Gaussian(N, Samples, sd, Radius)

sigma = .1*np.max(A) #Noise term scaled by curve height

A = A + sigma**2*numpy.random.random(A.shape)

func = X**3 #Function to estimate

V,S,U = np.linalg.svd(A)

C = np.dot(V.T,func) #Project the function onto the A basis

C_zeroed = np.array(C)
C_zeroed[N:]=0
funcCest = np.dot(V,C_zeroed)

fig, axes = subplots(1,2,figsize = (12,4))
#plt.plot(V.T[110:120,:].T)
axes[0].plot(C_zeroed)
axes[1].plot(funcCest)
axes[1].plot(func)

E_noise = np.dot(V[:,0:N],(Samples*sigma**2/(S**2 + Samples*sigma**2))*C[0:N])

errorC = np.sqrt((func-funcCest)**2)

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].plot(errorC)
axes[1].plot(np.sqrt(E_noise**2))
axes[1].set_ylim([0,.12])

Gamma = np.dot(A.T,A)
Sinv = np.array(S**2)
Sinv[N:]=0
Sinv[Sinv>0]=1.0/Sinv[Sinv>0]
GammaInv = np.dot(np.dot(U.T,np.diag(Sinv)),U)
decoders = np.dot(np.dot(GammaInv,A.T),func)
#decoders = U.T inv(S) V.T

funcEst = np.dot(A, decoders)
errorDec = np.sqrt((func-funcEst)**2)

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].plot(errorDec)
axes[1].plot(funcEst)
axes[1].plot(func)
```

## Another pure analytic version

```python
sd = .05 #SD of gaussians
N = 200 #Number of neurons
Samples = 500 #Number of samples of gaussian
sigma = .1 #Noise term
Radius = 1 #-1,1
noise = True

X, A = Gaussian(N, Samples, sd, Radius)

if noise == True:
    sigma = .1*np.max(A) #Noise term scaled by curve height
else:
    sigma = 0

A = A + sigma**2*numpy.random.random(A.shape)

#func = X**3 #Function to estimate
func = X*(X-1)*(X+1)

V,S,U = np.linalg.svd(A)

C = np.fft.fft(func) #Project the function onto the A basis
spec = np.real(np.fft.fftshift(C))
spec = np.abs(spec)**2

#Compute the estimate with fourier basis
num_C = 60
C_static = np.array(C)
C_static[num_C/2:Samples-num_C/2]=0
C_static[0:num_C/2]=1
C_static[Samples-num_C/2:]=1
C_static = C_static*C  #Mask C
func_est = np.real(np.fft.ifft(C_static))

#Compute decoders from S
Sinv = np.array(S)
Sinv[num_C:]=0  #Use as many as for fourier
Sinv[Sinv>0]=1.0/Sinv[Sinv>0]
Uinv = np.dot(U.T,np.diag(Sinv))
Vf = np.dot(V.T,func)
decoders = np.dot(Uinv,Vf[0:N])

dec_est = np.dot(A, decoders)

error = np.sqrt((func-dec_est)**2) #Decoder error

E_static = np.sqrt((func-func_est)**2) #Analytic error
E_noise = np.dot(V[:,0:N],(Samples*sigma**2/((S**2)[0:N] + Samples*sigma**2))*C[0:N])
E_theory = np.sqrt((E_static+E_noise)**2)
```

```python
#Plots
fig, axes = subplots(1,3,figsize=(12,4))
axes[0].plot(error)
axes[1].plot(E_theory)
axes[2].plot(E_static)
```
