```python
import numpy as np
import matplotlib.pyplot as plt    
%pylab inline
```

```python
#Define rectified linear neuron rate function (scalars only)
def Gaussian(num, xSize, sigma):
    #xRange=[-.95,.95] #Smaller than diameter (i.e 2*R)
    Radius = 1
    x = linspace(-Radius ,Radius, xSize)
    means = linspace(-Radius, Radius, num)
    
    A = exp(-(x[:,None]-means[:])**2/(2*sigma**2))
    A = 1/(np.sqrt(2*np.pi*sigma**2)) * A
    return x, A
```

```python
BasisSize = 5 #Number of basis fcns to plot
sigma = .05 #SD of gaussians
N = 200 #Number of neurons
Samples = 1000 #Number of samples of gaussian
noise = True

X, A = Gaussian(N, Samples, sigma)

if noise == True:
    sigmaN = .1*np.max(A) #Noise term scaled by curve height
else:
    sigmaN = 0
A = A + sigmaN**2*numpy.random.random(A.shape)

G = np.dot(A.T,A) 
U, S, V = np.linalg.svd(G)
Chi = np.dot(U.T,A.T)
Chi = Chi.T/Chi.max(axis=1) #Normalize to 1

fig, axes = plt.subplots(1, 2, figsize=(12,3))
axes[0].plot(Chi[:,0:BasisSize])
axes[0].set_title('Bases')
axes[1].plot(X,A[:,0::10])
axes[1].set_title('Sample Tuning Curves')

np.sum(A[:,50])*2.0/Samples #Check normalization
```

```python
#Compare singular values from two sources
fig, axes = plt.subplots(1, 2, figsize=(12,3))
axes[0].plot(S[0:50])

T = np.dot(A,A.T) #+ N*sigmaN**2
S2,v = np.linalg.eig(T)
axes[0].semilogy(S2[0:50])
axes[0].legend(['svd(A.T,A)','svd(A,A.T'])

#See how 'toeplitz' the matrix looks
axes[1].matshow(T);
```

```python
def ComputeTk(k, sigma, delta, N, noise):
    sd = sigma/(np.sqrt(2)*delta)
    norm_factor = N/(4*delta*np.sqrt(2*np.pi)*sd)
    tk = norm_factor*np.exp(-k**2/(2*sd**2)) + N*noise**2
    return tk
```

```python
#Compare the theory and numerics for the diagonal
print ComputeTk(0, sigma, 1.0/Samples, N, sigmaN)
print np.max(diag(T))

Tk = []
for k in range(0, Samples): 
    Tk.append(ComputeTk(k-Samples/2.0, sigma, 1.0/Samples,N, sigmaN))
```

```python
#Fourier transform of one row to estimate singular values
row = T[Samples/2,:]
sv_est = np.real(np.fft.fft(np.fft.fftshift(Tk)))
sv_est = sv_est + (N/2)*sigmaN**2

eigenvs,v = np.linalg.eig(T)

fig,axes = subplots(1,3,figsize=(16,4))
axes[0].plot(row)
axes[0].plot(Tk)
axes[0].legend(['Row from matrix','Analytical estimate'],loc=3)

axes[1].plot(eigenvs[0:60:2],'r') #Step by two because of different indexing
axes[1].plot(sv_est[0:30])
axes[1].legend(['Eigen Values', 'FFT estimate'],loc=1)

axes[2].semilogy(eigenvs[0:60:2],'r') #Step by two because of different indexing
axes[2].semilogy(sv_est[0:30])
axes[2].legend(['Eigen Values', 'FFT estimate'],loc=1)
```
