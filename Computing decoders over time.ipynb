This code is for computing decoders using continuous input signals into spiking neurons. This is to test a method that should be good for generating decoders for neurogrid using continuously changing, real time input.

```python
import neurogrid as ng
from numpy import *
from matplotlib.pyplot import *
```

```python
def genSignal(T, rng, dt=0.001, rms=.5, bandwidth=[0*pi, 10*pi]):
    
    N = 2*floor(T/(2*dt))+1
    domega = 2*pi/T #radians divisions
    omega = domega*arange(0,(N-1)/2) #radians axis
    
    N = omega.shape[0]
    Amps = zeros(N) + 1j*zeros(N)
    
    index = logical_and((abs(omega)<=bandwidth[1]),(abs(omega)>=bandwidth[0])) #find the part of the axis we're interested in
    num = index.shape[0]
    
    Amps[index] = (rng.randn(1,num) + 1j*rng.randn(1,num)) # generates guassian white noise coefficients 
    
    Amps = Amps[None,:]
    ConjAmps = Amps.conj()
    Amps = concatenate((fliplr(ConjAmps[:,1:]), Amps), axis=1)
    
    S = ifft(ifftshift(Amps)).real #get time domain signal
    
    # scale with rms
    rmsS = sqrt(sum(S**2)/S.shape[1])
    S = S*(rms/rmsS)
    
    return S
```

```python
#Test the signal generator
rng = random.RandomState()
Sig = genSignal(1., rng)
plot(Sig.T,'b')
disp('Signal')
```

```python
rng = random.RandomState()

N = 20
Samples = 300 #Can't be bigger than the whole signal
n = ng.neurons.SpikeNeuron(N, rng, balanced=True, nonlinear=0)
e = ng.encoders.random(N, 1, rng)

X, A = ng.activity.classic(n, e, rng, use_spikes=True, X=Sig[:,0:Samples].T, sample_count=Samples)
X, A2 = ng.activity.classic(n, e, rng, use_spikes=False, X=Sig[:,0:Samples].T, sample_count=Samples)
disp('Done')
```

```python
plot(X[:,0],A.T)
plot(X[:,0],A2.T)
axis('tight')
disp('Rate vs Spike curves')
```

```python
N = 10
D = 1
T = 10
dt = 0.001
pstc = 0.01

rng = random.RandomState()
Sig = genSignal(T, rng, dt=dt, bandwidth=[0*pi,1*pi])

decay = exp(-dt/pstc)

a = ng.ensemble.Ensemble(N, N, D, seed=None)

f_max = 1000
inputs = outer(a.encoders,Sig.T)*f_max   
e_input = where(inputs>0, inputs, 0)
i_input = where(inputs<0, -inputs, 0)

fs = zeros(N*N, dtype='f')

A = []
for j in arange(Sig.shape[1]):
    s = a.neurons.tick(e_input[:,j], i_input[:,j], dt)
    fs = fs * decay + s * (1-decay)
    
    A.append(fs)

A = array(A)
G = dot(A.T, A)
U = dot(A.T, Sig[0:1,0:A.shape[0]].T)

Ginv = linalg.pinv(G, rcond=0.0001)
d = dot(Ginv, U)
```

```python
#Test with a new signal
T2=1.
pstc = 0.01

decay = exp(-dt/pstc)

SigTest = genSignal(T2,rng,bandwidth=[0*pi,20*pi])
inputs = outer(a.encoders,SigTest.T)*f_max   
e_input = where(inputs>0, inputs, 0)
i_input = where(inputs<0, -inputs, 0)

fs = zeros(N*N, dtype='f')

Xest = []
for j in arange(SigTest.shape[1]):
    s = a.neurons.tick(e_input[:,j], i_input[:,j], dt)
    fs = fs * decay + s * (1-decay)
    
    Xhat = dot(fs, d)
    
    Xest.append(Xhat)

Xest = array(Xest)

#rmse = np.sqrt(np.sum((X_vals[indices]-Y[indices])**2)/S)
#print rmse
```

```python
plot(Xest)
plot(SigTest.T)
disp('done')
```

```python
import scipy as sp

pstc = 0.01
    
#filter the spike data
def spikeFilter(spikes, pstc):
    dt = 0.001
    PSCfilter = exp(-arange(0.,1.,dt)/pstc)  #1 second filter
    PSCfilter = PSCfilter/(np.sum(PSCfilter)*dt) #Normalize
    
    A = sp.signal.fftconvolve(spikes,PSCfilter[:,None])
    return A[0:spikes.shape[0],:]

A = spikeFilter(spikes, pstc)
A1 = spikeFilter(spikes1, pstc)
A2 = spikeFilter(spikes2, pstc)
```
