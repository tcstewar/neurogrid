This notebook might be easier to read if the Function quality notebook is read first. There we showed that $MSE \propto N^{kP}/N$ where $N$ is the number of neurons, $P$ is the (largest) power of a polynomial function and $k$ is a function of the tuning curves $a_i$.

Here we are interested in the same kinds of effects when using sparsification techniques.

```python
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,11,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 20
noise=0.1 #Standard deviation of expected noise

Xlen = 100 #Number of sample points
xi=linspace(-1,1,Xlen)
basisSize = 7 #Number of basis fcns
LegNum = arange(0,basisSize)
LegCoeffMat = eye(basisSize)
LegBasis=zeros([basisSize, Xlen])
for i in LegNum:
    LegBasis[i,:] = np.polynomial.legendre.legval(xi,LegCoeffMat[i,:])

numPowers=arange(6)
    
PavgMSE=[]
for P in numPowers:
    avgMSE=[]
    for N in numNeurons:
        MSE=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=Xlen)
            #d = ng.decoders.sparse_greedy(A, X**P, rng, noise=noise, sparsity=0,minSV=0)
            d = ng.decoders.sparse_greedy(A, LegBasis[P,:][:,None], rng, noise=noise, minSV=0)

            A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise
            #A_noise = A  #Use this if you want to see decoding under no noise
            Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
            #MSE.append(mean((X**P-Xhat)**2))  #Compute MSE
            MSE.append(mean((LegBasis[P,:][:,None]-Xhat)**2))  #Compute MSE
            
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt
scale=.8 #For scaling the plotted comparisons e.g., 1/N
tmp=plt.plot(array(PavgMSE).T)
plt.legend(tmp,LegNum, loc=3)
plt.yscale('log')
plt.xticks(arange(0,11,1),array(numNeurons).astype('str'))
plt.plot(scale/numNeurons,'--')
plt.plot(scale/numNeurons**.5,'.')
disp('Accuracy for Pth Degree Legendre Polynomials with increasing neurons under 80% sparsity')
```

```python
data = array(PavgMSE)[:,6:] #Start fit at N=64, for large noise need to increase N before you get anything useful
length = data.shape[1]

xi = arange(0,length)
B = array([xi, ones(length)])
dataLog = log(data)  #Do it in log space since N was log

w=[]
for row in dataLog:
    w.append(linalg.lstsq(B.T, row)[0]) # obtaining the parameters
    
w=array(w)
plot(outer(w[:,0],xi).T+w[:,1])
plot(dataLog.T,'.')
disp('Data with linear fits')

#Linear fit to slopes
length = w.shape[0]
xi = arange(0,length)
B = array([xi, ones(length)])

w2 = linalg.lstsq(B.T, w[:,0])[0]

figure(2);
plot(w2[0]*xi+w2[1])
plot(w[:,0],'.')
disp(w[:,0])
disp(w2[0])
```

So first appearances are that we still have $MSE \propto N^{kP}/N$ and $k=0$ is about the same.

Let's do some direct comparisons using dense and sparse decoders doing the same thing on the same graphs.

```python
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,11,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 20
noise=0.1 #Standard deviation of expected noise

Xlen = 100 #Number of sample points
xi=linspace(-1,1,Xlen)
basisSize = 7 #Number of basis fcns
LegNum = arange(0,basisSize)
LegCoeffMat = eye(basisSize)
LegBasis=zeros([basisSize, Xlen])
for i in LegNum:
    LegBasis[i,:] = np.polynomial.legendre.legval(xi,LegCoeffMat[i,:])

PavgMSE=[]
PavgMSE_sparse=[]
for P in LegNum:
    avgMSE=[]
    avgMSE_sparse=[]
    for N in numNeurons:
        MSE=[]
        MSE_sparse=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=Xlen) #Standard NEF decoders
            d = ng.decoders.classic2(A, LegBasis[P,:][:,None], rng, noise=noise, minSV=0)
            d_sparse = ng.decoders.sparse_greedy(A, LegBasis[P,:][:,None], rng, noise=noise, minSV=0)

            A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise
            #A_noise = A  #Use this if you want to see decoding under no noise
            Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
            Xhat_sparse = np.dot(A_noise.T, d_sparse)     #Compute estimate with decoders
            MSE.append(mean((LegBasis[P,:][:,None]-Xhat)**2))  #Compute MSE
            MSE_sparse.append(mean((LegBasis[P,:][:,None]-Xhat_sparse)**2))  #Compute MSE
            
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
        avgMSE_sparse.append(mean(MSE_sparse))
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
    PavgMSE_sparse.append(avgMSE_sparse)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt
scale=.8 #For scaling the plotted comparisons e.g., 1/N
ax = gca()
ax.set_color_cycle(['b','g','r','m','k','y', 'b'])
tmp=plot(array(PavgMSE).T)
plot(array(PavgMSE_sparse).T,'--')

legend(tmp,LegNum, loc=3)

yscale('log')
xticks(arange(0,11,1),array(numNeurons).astype('str'))
plot(scale/numNeurons,'.')
plot(scale/numNeurons**2,'+')
disp('Accuracy for Nth Degree Polynomials sparse vs non-sparse decoders')
```

This graph seems to support the hunch that sparsity is generally shifting all MSEs up, though clearly not by constant amounts. By the time you're over the first few powers (i.e. after $x^3$) they are quite similar in $\delta MSE/neuron$.  

This is for 80% sparsity.  Let's see the effects of varying the sparsity parameter.

```python
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,10,1) #2,4,8...  make 12 ~10 to run faster
sparseList = arange(.5,1,.05)
numRuns = 20

noise=0.1 #Standard deviation of expected noise

P=4  #The polynomial to test sparsity on

Xlen = 100 #Number of sample points
xi=linspace(-1,1,Xlen)
LegCoeffMat=zeros(P+1)
LegCoeffMat[P]=1
LegBasis = np.polynomial.legendre.legval(xi,LegCoeffMat)

SavgMSE_sparse=[]
for sparsity in sparseList:
    avgMSE_sparse=[]
    for N in numNeurons:
        MSE_sparse=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=100) #Standard NEF decoders
            d_sparse = ng.decoders.sparse_greedy(A, X**P, rng, noise=noise, sparsity=sparsity, minSV=0)

            A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise
            #A_noise = A  #Use this if you want to see decoding under no noise
            Xhat_sparse = np.dot(A_noise.T, d_sparse)     #Compute estimate with decoders
            MSE_sparse.append(mean((X**P-Xhat_sparse)**2))  #Compute MSE
            
        avgMSE_sparse.append(mean(MSE_sparse))
    SavgMSE_sparse.append(avgMSE_sparse)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt
scale=.8 #For scaling the plotted comparisons e.g., 1/N
fig3 = figure()
tmp=plot(array(SavgMSE_sparse).T)
legend(tmp,sparseList, loc=3)
title('P=%i' % P)

yscale('log')
xticks(arange(0,11,1),array(numNeurons).astype('str'))
plot(scale/numNeurons,'.')
plot(scale/numNeurons**2,'+')
disp('Accuracy for increasing neurons with varying sparse decoders')
```

Running the above code, it's clear that sparsity has almost no effect until about 0.7 for $P>1$.  That is why the plots above start at 0.6 for $P>1$.  After this point, it seems that <i>lower</i> $P$ are worse represented at high sparsity.  There is an over all shift up for higher $P$, but there is a $1/N$ drop in MSE even at high $P$. This makes it being the case that at $sparsity=.95$, 4th order polynomials are represented better than 2nd order for $N>64$.
