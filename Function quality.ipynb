# Question: 

What is the relation between the number of neurons and the quality of function computation?

<li>Assume an even distribution of tuning curves (as this is found in neurogrid)</li>
<li> The answer depends on the specific f(x) being computed and its relation to singular values of the population (SV)</li>

## More specific question:

What is the relation between the number of neurons and any given singular value?


We actually know one answer, namely in the case of the second SV (i.e. for linear functions).

### Linear Functions

Rate neurons and the accuracy of representation:<p>
$MSE = \frac{1}{2}\int\[x-\hat{x}\]^2dx + \frac{1}{2}\sigma^2 \sum_i d_i^2$

First term is distortion error. Second term is noise error.  The first one is small, so it comes down to the noise.
Meaning,
$$MSE\propto\sum_i d_i^2 \;\;\; (1)$$
Also, recall that 
$$\hat{x} = \sum_i a_i(x) d_i$$
Now, here are two bad derivations that suggestion the same thing


#### Derivation 1


Assume,
$$1= \sum_i a_i(1) d_i$$
So $a(1)=R$ constant max firing. Thus,
$$1 = RNd_i/2$$
$$d_i=\frac{2}{NR}$$
So,
$$d_i\propto \frac{1}{N}$$ since $R$ and 2 are constant.



#### Derivation 2

Assume the tuning curves are linear, so we have
$$x=\sum_i m_i x d_i + b_i d_i$$
$$=\sum_i m_i x d_i + \sum_i b_i d_i$$
So
$$x - \sum_i b_i d_i \sum_i m_i x d_i$$
$$1 - N [b_i] [d_i]/x = N [m_i] [d_i]$$
Note: I'm using $[m_i]$ to indicate the mean since angle brackets don't work in iPython Notebook.
Dividing through gives
$$1/N - [b_i] [d_i]/x = [m_i] [d_i]$$
Because all of the means are constant, we can see that for all values of $x$
$$[d_i] \propto \frac{1}{N}$$

So, using either derivation, and from (1) we have:
$$MSE\propto\sum_i \frac{1}{N^2} = N/N^2 = 1/N$$


Multiplying $x$ by a constant (i.e. having a generic linear function) won't change these derivations.  Let's check the result with code.

```python
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,11,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 50
noise=0.1 #Standard deviation of expected noise
scale=0.3 #For scaling the plotted comparisons e.g., 1/N

avgMSE=[]
for N in numNeurons:
    MSE=[]
    for i in arange(numRuns):
        n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
        e = ng.encoders.random(N, 1, rng)
        
        X, A = ng.activity.classic(n, e, rng, sample_count=100) #Standard NEF decoders
        d = ng.decoders.classic2(A, X, rng, noise=noise, minSV=0)
        A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise

        Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
        MSE.append(mean((X-Xhat)**2))  #Compute MSE
    
    avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
        
import matplotlib.pyplot as plt    
plt.plot(avgMSE, label='MSE')
plt.plot(scale/numNeurons,'r', label='1/N')
plt.yscale('log')
plt.legend()
plt.xticks(arange(0,11,1),array(numNeurons).astype('str'))
disp('Representation Error as N increases')
```

Here's the same approximation for a population of 1000 rectified linear neurons.

```python
#Define rectified linear neuron rate function (scalars only)
def rectLin(num, xSize):
    maxRange=[100,200]
    xRange=[-.95,.95] #Smaller than diameter (i.e 2*R)
    Radius = 1
    x = linspace(-Radius,Radius,xSize)
    
    xInts = (xRange[1]-xRange[0])*rng.rand(1,num)+xRange[0]  #Generate random x intercepts
    maxFRs = (maxRange[1]-maxRange[0])*rng.rand(1,num)+maxRange[0]  #Generate random max firing rates
    slopes = maxFRs/(Radius-xInts)  #Compute the slopes
    biases = -(slopes*xInts)
    
    slopes[0,0:floor(num/2.0)]=-slopes[0,0:floor(num/2.0)] #Flip half the neurons
    
    A = outer(slopes,x).T+biases
    return x, A.clip(min=0)
```

```python
BasisSize = 5 #Number of basis fcns to plot
X, A = rectLin(1000,1500)
A = A.T
G = np.dot(A,A.T)
U,S,V=np.linalg.svd(G)
Chi = np.dot(U.T,A)
Chi = Chi.T/Chi.max(axis=1) #Normalize to 1
plt.plot(Chi[:,0:BasisSize])
disp('Bases')
```

## Relationship between MSE, polynomial order and number of neurons

Here is code demonstrating the relationship between order of Legendre polynomial, number of neurons and MSE numerically.

```python
#Find the relationship between Legendre polynomials and MSE
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,11,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 20
noise=0.1 #Standard deviation of expected noise

Xlen = 100 #Number of sample points
xi = linspace(-1,1,Xlen)
basisSize = 8 #Number of basis fcns
LegNum = arange(0,basisSize)
LegCoeffMat = eye(basisSize)
LegBasis=zeros([basisSize, Xlen])
for i in LegNum:
    LegBasis[i,:] = np.polynomial.legendre.legval(xi,LegCoeffMat[i,:])

PavgMSE=[]
for P in LegNum:
    avgMSE=[]
    for N in numNeurons:
        MSE=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=Xlen) #Standard NEF decoders
            d = ng.decoders.classic2(A, LegBasis[P,:], rng, noise=noise, minSV=0)

            A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise
            #A_noise = A  #Use this if you want to see decoding under no noise
            Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
            MSE.append(mean((LegBasis[P,:]-Xhat)**2))  #Compute MSE
        
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt

tmp=plt.plot(array(PavgMSE).T)
plt.legend(tmp,LegNum, loc=3)
plt.yscale('log')
plt.xticks(log2(numNeurons),array(numNeurons).astype('str'))
plt.plot(1.0/numNeurons,'--')
disp('Accuracy for Nth Degree Legendre Polynomials with increasing neurons')
```

Let's do linear regression to see the slopes of the lines of best fit for this data, order to see how the MSE changes with polynomial order.

```python
data = array(PavgMSE)[:,6:] #Start fit at N=64
length = data.shape[1]
```

```python
xi = arange(0,length)
A = array([xi, ones(length)])
dataLog = log(data)

w=[]
for row in dataLog:
    w.append(linalg.lstsq(A.T, row)[0]) # obtaining the parameters
```

```python
w=array(w)
plot(outer(w[:,0],xi).T+w[:,1])
plot(dataLog.T,'.')
disp('Data with linear fits')
```

```python
#Linear fit to slopes
length = w.shape[0]
xi = arange(0,length)
A = array([xi, ones(length)])

w2 = linalg.lstsq(A.T, w[:,0])[0]

plot(w2[0]*xi+w2[1])
plot(w[:,0],'.')
disp(w2[0])
```

Here, the slope is some small value (printed above the graph), which suggests a $k$ for $MSE \propto N^{kP}/N$.  This $k$ depends on the $a_i$ used (see below).

### Formula for computing N

Here's a question we might want answered in general:
How many neurons do I need if I want to compute a function $f(x)$ with a given $MSE$ under a given amount of noise $\sigma^2$?

To answer this we must determine all of the scaling and bias factors left out of the previous analysis.  Note that we implicitly accounted for noise (usually left at 0.1) in that analysis.  The following does the same because, after consideration, it seems likely that we'll know the amount of noise and that it will stay constant. So the best way to use this code is run it for the desired $\sigma$, and then use the resulting theoretical predictions (i.e. the code computes the constants for the theoretical fits for a given amount of noise).  The result will provide an answer to the question: 
<p>How many neurons are needed to compute the function $x^P$ to a certain MSE under a given amount of noise $\sigma^2$?</p>

Recall that we did a linear fit to changes in slope as a function of P. We can do that again here.

```python
#Find the relationship between Legendre polynomials and MSE
import neurogrid as ng
import numpy as np

rng = np.random.RandomState()
numNeurons=2**arange(0,9,1) #2,4,8...  make 12 ~10 to run faster
numRuns = 10
noise=0.1 #Standard deviation of expected noise

Xlen = 100 #Number of sample points
xi=linspace(-1,1,Xlen)
basisSize = 7 #Number of basis fcns
LegNum = arange(0,basisSize)
LegCoeffMat = eye(basisSize)
LegBasis=zeros([basisSize, Xlen])
for i in LegNum:
    LegBasis[i,:] = np.polynomial.legendre.legval(xi,LegCoeffMat[i,:])

PavgMSE=[]
for P in LegNum:
    avgMSE=[]
    for N in numNeurons:
        MSE=[]
        for i in arange(numRuns):
            n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
            e = ng.encoders.random(N, 1, rng)
            
            X, A = ng.activity.classic(n, e, rng, sample_count=Xlen) #Standard NEF decoders
            d = ng.decoders.classic2(A, LegBasis[P,:], rng, noise=noise, minSV=0)

            A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise
            #A_noise = A  #Use this if you want to see decoding under no noise
            Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
            MSE.append(mean((LegBasis[P,:]-Xhat)**2))  #Compute MSE
        
        avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
    PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
disp('Done')
```

```python
import matplotlib.pyplot as plt

tmp=plt.plot(array(PavgMSE).T)
plt.legend(tmp,LegNum, loc=3)
plt.yscale('log')
plt.xticks(log2(numNeurons),array(numNeurons).astype('str'))
plt.plot(1.0/numNeurons,'--')
disp('Accuracy for Nth Degree Legendre Polynomials with increasing neurons')
```

And now the linear fitting for the slopes.

```python
data = array(PavgMSE)[:,6:] #Start fit at N=64, for large noise need to increase N before you get anything useful
length = data.shape[1]

xi = arange(0,length)
A = array([xi, ones(length)])
dataLog = log(data)

w=[]
for row in dataLog:
    w.append(linalg.lstsq(A.T, row)[0]) # obtaining the parameters
    
w=array(w)
plot(outer(w[:,0],xi).T+w[:,1])
plot(dataLog.T,'.')
disp('Data with linear fits')

#Fit to slopes... start linear
length = w.shape[0]
xi = arange(0,length)
A = array([xi, ones(length)])

optSlope = linalg.lstsq(A.T, w[:,0])[0]

figure(2);
plot(optSlope[0]*xi+optSlope[1])
plot(w[:,0],'.')

disp(w[:,0])
disp(optSlope)

```

Since we now care about the intercepts as well, we can try a linear fit for the intercepts. 

```python
#Linear fit to intercepts
length = w.shape[0]
xi = arange(0,length)
A = array([xi, ones(length)])

w3 = linalg.lstsq(A.T, w[:,1])[0]

figure(2)
plot(w3[0]*xi+w3[1])
plot(w[:,1],'.')

#That doesn't work well, so do a nonlinear fit 
from scipy import optimize
fitfunc = lambda p, x: p[0]*p[2]**(-x) - p[1] 
errfunc = lambda p, x, y: fitfunc(p, x) - y
p0 = [-6.0,-1.0,2.0] #Start guess
optInter,success = optimize.leastsq(errfunc, p0[:], args=(xi, w[:,1]))

plot(w[:,1],'.')
plot(optInter[0]*optInter[2]**(-xi) - optInter[1] )

disp(optInter)
```

So the linear fit to intercepts didn't work, but the fit of the nonlinear function $C_1 C_3^{-x} - C_2$ works well.

Now that we have fits for the slopes and intercepts (recall, from 64 neurons on), we can use these to fit all the data.

```python
#Simplifying the previous expression
figure()
ax = gca()
ax.set_color_cycle(['b','g','r','m','k','y','c'])

P=arange(0,7)

#testing predictions
f1 = optSlope[0]*P + optSlope[1]
f2 = optInter[0]*optInter[2]**(-P) - optInter[1] 
f3 = log2(numNeurons) - 6  #The value 6 comes from the fact that we started the fit at 2^6 neurons.
prediction = exp(outer(f1,f3).T+f2)

toPlot = array(PavgMSE)

tmp=plt.loglog(numNeurons, toPlot.T)
plt.legend(tmp,P, loc=3)
plt.loglog(numNeurons, prediction,'--')
axis('tight')
disp('Data versus theory Legendre')
disp(optSlope)
disp(optInter)
```

Doing some algebra gave those expressions for the functions that are included in the fit.  Specifically,

$\begin{eqnarray}
MSE = g(P,N,\sigma) = exp(f_2(P) + f_1(P)f_3(N))
\end{eqnarray}$

where

$\begin{eqnarray}
f_1(P) &=& C_1P-C_2 \\\
f_2(P) &=& C_3 2^{-P} -1 \\\
f_3(N) &=& log_2(N)-C_4
\end{eqnarray}$


and, from the data fits we have the values of $C_i$.


So, we now have the answer to our question: How many neurons are needed to compute the $P^{th}$ order Legendre basis to a certain MSE under a given amount of noise $\sigma^2$?

$$N = 2^{f(P,MSE,\sigma)}$$

where

$$f(P,MSE,\sigma^2) = (log(MSE)-f_2(P))/f_1(P)-C_4$$

We are now in a position to define the function that we want to use to determine the number of neurons needed to reach a given MSE for any function we are trying to compute, $f(x)$.  

First we convert the function to a representation on a Legendre basis:

$\begin{eqnarray}
f(x) &=& \sum_i^P C_i L_i(x) \\\
C_i &=& \int f(x) L_i(x) dx
\end{eqnarray}$

If we assume that the MSEs between basis functions are independent (which they technically are not, but they are close), then we have
$$MSE = \sum_i^P C_i g_i(N,\sigma) $$
and we want to find the minimum $N$ to meet this MSE.  Each $g_i(N,\sigma)$ has already been found, so we can use a nonlinear function minimizer to get the answer.  The following code demonstrates the full process, going from some function $f(x)$ to determining how many neurons will be needed to reach a given MSE assuming a certain $\sigma^2$.

```python
from scipy import optimize
import neurogrid as ng
import numpy as np

#Find the relationship between Legendre polynomials and MSE for neurogrid neurons
#This code encapsulates the previous work in a function and is called by
#the optimizer below
def MSEPredictorCoefficients(noise, basisSize, startNeurons, sparsity=None):
    rng = np.random.RandomState()
    numNeurons=2**arange(startNeurons,11,1) #2,4,8...  make 12 ~10 to run faster
    numRuns = 20
    
    Xlen = 100 #Number of sample points need a large number for nnls
    xi=linspace(-1,1,Xlen)
    LegNum = arange(0,basisSize)
    LegCoeffMat = eye(basisSize)
    LegBasis=zeros([basisSize, Xlen])
    for i in LegNum:
        LegBasis[i,:] = np.polynomial.legendre.legval(xi,LegCoeffMat[i,:])
        
    PavgMSE=[]
    for P in LegNum:
        avgMSE=[]
        for N in numNeurons:
            MSE=[]
            for i in arange(numRuns):
                n = ng.neurons.RateNeuron(N, rng, balanced=True, nonlinear=0)
                e = ng.encoders.random(N, 1, rng)
                
                X, A = ng.activity.classic(n, e, rng, sample_count=Xlen) 
                if sparsity is None:
                    d = ng.decoders.classic2(A, LegBasis[P,:], rng, noise=noise, minSV=0)
                else:
                    d = ng.decoders.sparse_greedy(A, LegBasis[P,:][:,None], rng, noise=noise, minSV=0, sparsity=sparsity)
    
                A_noise = A + rng.randn(*A.shape)*(noise*np.max(A)) #Test under noise

                Xhat = np.dot(A_noise.T, d)     #Compute estimate with decoders
                MSE.append(mean((LegBasis[P,:]-Xhat)**2))  #Compute MSE
            
            avgMSE.append(mean(MSE))  #Take the mean MSE over numRuns
        PavgMSE.append(avgMSE)  #Collect the MSE across powers
        
    data = array(PavgMSE)[:,0:] #Start fit at N=64, for large noise need to increase N before you get anything useful
    length = data.shape[1]
    
    xi = arange(0,length)
    A = array([xi, ones(length)])
    dataLog = log(data)
    
    w=[]
    for row in dataLog:
        w.append(linalg.lstsq(A.T, row)[0]) # obtaining the parameters
        
    w=array(w)
    
    #Fit to slopes... linear
    length = w.shape[0]
    xi = arange(0,length)
    A = array([xi, ones(length)])
    
    optSlope = linalg.lstsq(A.T, w[:,0])[0]
    
    #Fit to intercepts... nonlinear
    fitfunc = lambda p, x: p[0]*p[2]**(-x) - p[1] 
    errfunc = lambda p, x, y: fitfunc(p, x) - y
    p0 = [-6.0,-1.0,2.0] #Start guess
    optInter,success = optimize.leastsq(errfunc, p0[:], args=(xi, w[:,1]))

    return optSlope, optInter #The constants for the fit
```

```python
#First we find the constants in our known MSE functions for each legendre basis
#This basically calls a function that encapsulates the code developed above
#It also includes a sparsity option, which is described in the Function quality
#under sparsification notebook.
startNeurons = 6 #Power of 2 to start the coefficient generation at
noise = 0.1
basisSize = 7 #Number of legendre bases to use to generate the fit
optSlope, optInter = MSEPredictorCoefficients(noise,basisSize,startNeurons,sparsity=None)  #Really don't have to call this every time
```

```python
import neurogrid as ng
import numpy as np

#Define the problem by picking the function we're going to use, setting an MSE and choosing \sigma
xLen = 1000
x = linspace(-1,1,xLen)
dx = 2./xLen
MSE = 0.001
fx = x[:]**2

#Now we find the coefficients of the representation of this on the Legendre basis
basisMax = 10 #Max number of basis functions to use, only hit if the basis MSE isn't low enough
LegCoeffMat = eye(basisMax)
basisMSE = 0
i = 0
while i < basisMax:
    C=np.polynomial.legendre.legfit(x,fx,i)
    basisMSE = mean((np.polynomial.legendre.legval(x,C) - fx)**2)
    if basisMSE < MSE/10:
        break
    i += 1
```

```python
#We know the nonlinear functions to be called by the optimizer since we have found the
#predictive coefficients
def MSEPrediction(P, optSlope, optinter, startNeurons, numNeurons):
    f1 = optSlope[0]*P + optSlope[1]
    f2 = optInter[0]*optInter[2]**(-P) - optInter[1] 
    f3 = log2(numNeurons) - startNeurons
    prediction = exp(outer(f1,f3).T+f2)
    prediction.shape = ()
    return prediction

#Last we use an optimizer in scipy to find the minimum N to meet the desired MSE
#The function to return the current MSE as a function of the number of neurons (the parameters for the fit)
def fitfunc(p):
    #p are the parameters, here the number of neurons for each basis
    #x are the coefficients on those bases for this function
    MSETotal = 0
    for i in arange(C.shape[0]):
        MSETotal += C[i]*MSEPrediction(i, optSlope, optInter, startNeurons, p[i])
    return MSETotal
    
errfunc = lambda p: (fitfunc(p) - MSE)**2
    
p0 = ones(C.shape[0])*10 #Start guess, 10 neurons for each basis function
optNumNeurons = optimize.fmin(errfunc, p0[:])

print 'Note that these values are mean numbers needed to get the desired MSE.'
print 'The number of neurons by basis is', list(optNumNeurons)
print 'The total number of neurons is', ceil(sum(optNumNeurons)), 'ish.'
```
